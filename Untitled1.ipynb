```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Select only the numeric score colums
score_columns = ['math score','reading score','writing score']
df_scores = df[score_columns]

#compute correlation matrix
corr = df_scores.corr()

#plot correlation heatmap
plt.figure(figsize=(6,4))
sns.heatmap(corr,annot=True,cmap='coolwarm',fmt=".2f")
plt.title('Correlation Heatmap of student Scores')
plt.tight_layout()
plt.show()

```

```python
from sklearn.preprocessing import MinMaxScaler

#Select the score columns to normalaize
score_columns = ['math score','reading score','writing score']
df_scores = df[score_columns]

#Initialize the Scler
scaler = MinMaxScaler()

#Fit and transfrom the selected columns
df_normalized = pd.DataFrame(scaler.fit_transform(df_scores),columns=score_columns)
#
df_final = pd.concat([df.drop(columns=score_columns).reset_index(drop=True),
                      df_normalized.reset_index(drop=True)],axis=1)

#show the normalized result
print(df_final.head())
```

```python
from sklearn.preprocessing import StandardScaler

#Select the score columns to StandardScaler
score_columns = ['math score','reading score','writing score']
df_scores = df[score_columns]

#Initialize the Scler
scaler = StandardScaler()

#Fit and transfrom the selected columns
df_standardized = pd.DataFrame(scaler.fit_transform(df_scores),columns=score_columns)
#
df_final = pd.concat([df.drop(columns=score_columns).reset_index(drop=True),
                      df_standardized.reset_index(drop=True)],axis=1)

#show the standardized result
print(df_final.head())
```
